```python
from tkinter import *
import tkinter

# Declaration
window = Tk()
window.title("Pandemic Analytics")
window.geometry("700x800")

def cal_func():

    total_population = entry2.get()
    int_total_population = int(total_population)

    infected = entry3.get()
    int_infected = int(infected)

    death = entry4.get()
    int_death = int(death)

    ifr = (int_death / int_infected) * 100
    cmr = (int_death / int_total_population)
    
    
    
    window = Tk()
    window.title("Result")

    label1 = Label(window, font=("helvetica", 22, "bold"), text="Result" , bg = "grey" , fg = "black")
    label1.grid(row=7, columnspan=1)

    label2 = Label(window, font=("times", 14, "bold"), text="IFR : ", bg = "grey" , fg = "black")
    label2.grid(row=9, column=1)
    label3 = Label(window, text=ifr)
    label3.grid(row=9, column=2)

    label4 = Label(window, font=("times", 14, "bold"), text="CMR : ", bg = "grey" , fg = "black")
    label4.grid(row=10, column=1)
    label5 = Label(window, text=cmr)
    label5.grid(row=10, column=2)

# Listing out all the states in alphabetical order.    
options = ['Andaman and Nicobar', 'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 'Chandigarh', 'Chhattisgarh', 'Dadra and Nagar Haveli', 'Delhi', 'Goa', 'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Jharkhand', 'Karnataka', 'Kerala', 'Ladakh', 'Lakshadweep', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Odisha', 'Puducherry', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu', 'Telengana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal']
select = StringVar() 
select.set(options[0])

# Setting font and colour.
label1 = Label(window,font=("helvetica", 22, "bold"), text="Pandemic Analytics",bg = "red" , fg = "black")
label1.grid(row=1,columnspan = 1)

label2 = Label(window, text="Enter the State:",bg = "grey" , fg = "black")
label2.grid(row=2)
drop = OptionMenu(window,select,*options)
drop.grid(row=2,column=2)

label3 = Label(window, text="Total population:" , bg = "grey" , fg = "black")
label3.grid(row=3)
entry2 = Entry(window)
entry2.grid(row=3, column=2)

label4 = Label(window, text="Enter Infected:" , bg = "grey" , fg = "black")
label4.grid(row=4)
entry3 = Entry(window)
entry3.grid(row=4, column=2)

label5 = Label(window, text="Enter deaths:" , bg = "grey" , fg = "black")
label5.grid(row=5)
entry4 = Entry(window)
entry4.grid(row=5, column=2)


# Button for calculating the IFR and CMR.
button1 = Button(window, text="Calculate IFR And CMR:",command=cal_func , bg = "grey" , fg = "black")
button1.grid(column=2)

# Displaying the GUI on to the console.
window.mainloop()



```
